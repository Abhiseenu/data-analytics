```python
data.ffill(inplace=True)

data.fillna(0,inplace=True)
data
```

```python
print(data.duplicated().sum())
```

```python

data.drop_duplicates(inplace=True)
print(data.duplicated().sum())
data
```

```python

data['month'] = pd.to_datetime(data['month'], format='%b %Y')
```

```python
# Convert 'age_group' to a categorical type without triggering the warning
data.loc[:, 'age_group'] = pd.Categorical(data['age_group'])

# Verify the conversion
print(data['age_group'].cat.categories)  # Check the categories
print(data['age_group'].dtype)           # Check the data type
data
```

```python
# Convert columns to numeric, coercing errors to NaN
data['crude_vax_ir'] = pd.to_numeric(data['crude_vax_ir'], errors='coerce')
data['crude_unvax_ir'] = pd.to_numeric(data['crude_unvax_ir'], errors='coerce')
data['crude_irr'] = pd.to_numeric(data['crude_irr'], errors='coerce')
data['age_adj_vax_ir'] = pd.to_numeric(data['age_adj_vax_ir'], errors='coerce')
data['age_adj_unvax_ir'] = pd.to_numeric(data['age_adj_unvax_ir'], errors='coerce')
data['age_adj_irr'] = pd.to_numeric(data['age_adj_irr'], errors='coerce')
data['monthly_age_adj_vax_ir'] = pd.to_numeric(data['monthly_age_adj_vax_ir'], errors='coerce')
data['monthly_age_adj_unvax_ir'] = pd.to_numeric(data['monthly_age_adj_unvax_ir'], errors='coerce')
data['monthly_age_adj_irr'] = pd.to_numeric(data['monthly_age_adj_irr'], errors='coerce')
data['monthly_age_adj_mb_irr'] = pd.to_numeric(data['monthly_age_adj_mb_irr'], errors='coerce')
```

```python
data = data[['age_group', 'crude_vax_ir', 'crude_unvax_ir', 'crude_irr',
         'age_adj_vax_ir', 'age_adj_unvax_ir', 'age_adj_irr',
         'monthly_age_adj_vax_ir', 'monthly_age_adj_unvax_ir',
         'monthly_age_adj_irr', 'monthly_age_adj_mb_irr']]
```

```python
age_group_data = data.groupby('age_group').sum()
print(age_group_data)
```

```python
age_group_data['mortality_rate'] = (age_group_data['crude_irr'] / age_group_data['crude_vax_ir']) * 100
print(age_group_data)
```

```python
import matplotlib.pyplot as plt

# Plot crude incidence rates by age group
plt.figure(figsize=(6, 6))
age_group_data['crude_vax_ir'].plot(kind='bar', color='skyblue')
plt.title('Crude Incidence Rates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Crude Incidence Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot age-adjusted incidence rates by age group
plt.figure(figsize=(6, 6))
age_group_data['age_adj_vax_ir'].plot(kind='bar', color='salmon')
plt.title('Age-Adjusted Incidence Rates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Age-Adjusted Incidence Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot mortality rates by age group
plt.figure(figsize=(6, 6))
age_group_data['mortality_rate'].plot(kind='bar', color='lightgreen')
plt.title('Mortality Rates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mortality Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Calculate mean and standard deviation of mortality rates
mean_mortality = np.mean(age_group_data['mortality_rate'])
std_mortality = np.std(age_group_data['mortality_rate'])
print(f'Mean Mortality Rate: {mean_mortality:.2f}%')
print(f'Standard Deviation: {std_mortality:.2f}')
```
